# Problem Definition

# File naming convention

Field description:

1. Province_State - The name of the State within the USA.
2. Country_Region - The name of the Country (US).
3. Last_Update - The most recent date the file was pushed.
4. Lat - Latitude.
5. Long_ - Longitude.
6. Confirmed - Aggregated confirmed case count for the state.
7. Deaths - Aggregated Death case count for the state.
8. Recovered - Aggregated Recovered case count for the state.
9. Active - Aggregated confirmed cases that have not been resolved (Active = Confirmed - Recovered - Deaths).
10. FIPS - Federal Information Processing Standards code that uniquely identifies counties within the USA.
11. Incident_Rate - confirmed cases per 100,000 persons.
12. People_Tested - Total number of people who have been tested.
13. People_Hospitalized - Total number of people hospitalized.
14. Mortality_Rate - Number recorded deaths * 100/ Number confirmed cases.
15. UID - Unique Identifier for each row entry.
16. ISO3 - Officialy assigned country code identifiers.
17.  Testing_Rate - Total number of people tested per 100,000 persons.
18. Hospitalization_Rate - Total number of people hospitalized * 100/ Number of confirmed cases.

```python
# dividing the x varaibles features from y variable target (Mortality_Rate)
x=ds.drop(columns=["Mortality_Rate"])
y=ds[["Mortality_Rate"]]
```

# To check the distribution of skewness

```python
ds.skew()  # skewness is there
```

```python
# removing skewness by transforming data
from sklearn.preprocessing import power_transform
```

```python
x=power_transform(x,method='yeo-johnson')
```

```python
# scaling the dataset
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_t=sc.fit_transform(x)
```

finding best random state

```python
# importing our model libraries
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
```

```python
from sklearn.linear_model import LinearRegression
best_rstate=0
accuracy=0
for i in range(30,200):
    x_train,x_test,y_train,y_test=train_test_split(x_t,y,test_size=.25,random_state=i)
    mod=LinearRegression()
    mod.fit(x_train,y_train)
    predlr=mod.predict(x_test)
    tempaccu=r2_score(y_test,predlr)
    if(tempaccu>accuracy):
        accuracy=tempaccu
        best_rstate=i
        
print("Best Accuracy",accuracy*100, "Random state",best_rstate)        
```

```python
# breaking out input and output variable into train and test data
x_train,x_test,y_train,y_test=train_test_split(x_t,y,test_size=.22,random_state=93)
```

```python
print(x_train.shape , '\t'  , y_train.shape)
```

```python
print(x_test.shape , '\t'  , y_test.shape)
```

```python
#using algorithms in for loops
model=[LinearRegression(),Ridge(),AdaBoostRegressor(),RandomForestRegressor(),KNeighborsRegressor(),GradientBoostingRegressor()]
for m in model:
    m.fit(x_train,y_train)
    print("Score of " , m , "is" , m.score(x_train,y_train))
    y_pred=m.predict(x_test)
    print("Mean Absolute error  " , mean_absolute_error(y_test,y_pred))
    print("Mean Squared error \n",mean_squared_error(y_test,y_pred))
    print("Root mean Squared error is \n",np.sqrt(mean_squared_error(y_test,y_pred)))
    print("r2 score " , r2_score(y_test,y_pred))
    print("**********************************************************************\n")
```

# selecting LinearRegression as best model because of its good score and r2 score

# Hyperparameter tuning

```python
# LinearRegression is best performing model so finding its best parameter
from sklearn.model_selection import GridSearchCV
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_t, y, test_size=0.22,random_state =93 )
model = LinearRegression()
parameters = {'fit_intercept':[True,False], 'normalize':[True,False], 'copy_X':[True, False]}
grid = GridSearchCV(model,parameters, cv=5)
grid.fit(x_train, y_train)
grid.best_params_
```

```python
# using LinearRegression with best parameters
lr=LinearRegression(copy_X=True,fit_intercept=True,normalize=True)
lr.fit(x_train,y_train)
print(" Score is ",lr.score(x_train,y_train))
predlr=lr.predict(x_test)
print("Mean Absolute error  " , mean_absolute_error(y_test,predlr))
print("Mean Squared error \n",mean_squared_error(y_test,predlr))
print("Root mean Squared error is \n",np.sqrt(mean_squared_error(y_test,predlr)))
print("r2 score " , r2_score(y_test,predlr))
print("**********************************************************************\n")
```

```python
#plotting scatter between test data and predict data
plt.scatter(y_test,predlr)
plt.xlabel("y_test")
plt.ylabel("predicted y")
plt.title("Scatter plot between test and predict data ")
plt.show()
```
