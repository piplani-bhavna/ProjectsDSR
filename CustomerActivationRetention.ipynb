## Data Visualization

#### Univariate Analysis and Bivariate analysis

```python
plt.figure(figsize=(15,5))
sns.countplot(df["2 How old are you? "])
plt.title("Age Wise distribution of Online shopping \n \n1 - (Less than 20) \t 2 - (21-30) years \t 3- (31-40) years \t 4 - (41-50) years \t 5 - (51 and above)")
plt.show()  
```

#### Above countplot shows people of age between 31-40 shoppe more than others

```python
sns.displot(df["16 How 4 do you abandon (selecting an items and leaving without making payment) your shopping cart?                                                   "])
plt.title("How frequently do customer abadon their shopping cart \n \n\t 1 - Never \t  3 - Sometimes\t 4 - Frequently \t 5 - Very frequently ")
plt.show() # showing the maximum no of customer sometimes abandon their shopping cart
```

#### Above displot shows the majority of people abandon their shopping cart sometimes

```python
plt.figure(figsize=(15,5))
sns.countplot(df["6 How many times you have made an online purchase in the past 1 year?"])
plt.title("Age Wise distribution of past 1 year Online purchase \n \n 1. Less than 10 times \t 2. 11-20 times \t 3. 21-30 times \t 4. 31-40 times \t 5. 41 and above")
plt.show() 
```

#### Above countplot shows the maximum customers have done online purchase less than 10 times in a year

```python
plt.figure(figsize=(15,5))
sns.countplot(df["12 Which channel did you follow to arrive at your favorite online store for the first time?                                                                   "])
plt.title("Channel used for online store \n \n 1. Search Engine \t 3. Content marketing \t 4. Display adverts ")
plt.show() 
```

### Above countplot shows majority of customer follow search engine to arrive at your favourite online store

```python
sns.barplot(x='2 How old are you? ',y='5 Since How Long You are Shopping Online ?',hue='1Gender of respondent',data=df)
plt.title("\n Gender Wise Age and online shopping Distribution\n ")
plt.figure(figsize=(10, 10))
plt.show()
```

#### Above barplot shows male of age between 31-40 years have maximum no of years of shopping online 

```python
sns.displot(df["15 What is your preferred payment Option?                                                                                                                               "])
plt.title("1. Credit/Debit cards \t 2. Cash on delivery (CoD) \t 4. E-wallets (Paytm, Freecharge etc.)")
plt.show() # showing the maximum no of customer sometimes abandon their shopping cart
```

### Above displot shows that majority of customer prefer credit/debit cards for payment options

```python
plt.figure(figsize=(15,5))
sns.countplot(df["24 User friendly Interface of the website"])
plt.title(" User friendly Interface of the website \n \n 1. Strongly disagree \t 2. Dis-agree \T 3. Indifferent \t 4. Agree \t 5. Strongly agree")
plt.show() 
```

### Above countplot shows that mostly customer strongly agree with the user friendly interface of the website

```python
plt.figure(figsize=(15,5))
sns.countplot(df["47 Getting value for money spent"])
plt.title("Getting value for money spent \n \n 1. Strongly disagree \t 2. Dis-agree \T 3. Indifferent \t 4. Agree \t 5. Strongly agree")
plt.show() 
```

### Above countplot shows that majority of customer agree on getting value for money spent
