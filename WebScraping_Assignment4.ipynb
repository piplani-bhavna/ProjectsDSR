```python
from selenium import webdriver
import time
import requests
import shutil
import os
from webdriver_manager.chrome import ChromeDriverManager

option = webdriver.ChromeOptions()
option.headless=True
driver = webdriver.Chrome(ChromeDriverManager().install(), options=option)

url="https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)

time.sleep(3)
Rank=[]
Name=[]
Artist=[]
Upload_Date=[]
Views=[]


try:
    #scraping  videoname
    try:
        name=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[2]")
            
        for i in name:
            Name.append(i.text)
    except NoSuchElementException   as e:
        Name.append("Info Missing")
            
            #scraping rank
    try:
        rank=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[1]")
            
        for i in rank:
            Rank.append(i.text)
    except NoSuchElementException as e:
        Rank.append("Not mentioned ")
            
    try: # scraping artist
        artist=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[3]")
            
        for i in artist:
            Artist.append(i.text)
               
    except NoSuchElementException as e:
        Artist.append("Not mentioned ")    
            
    try: # scraping date of upload
        upload_date=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[5]")
            
        for i in upload_date:
            Upload_Date.append(i.text)
    except NoSuchElementException as e:
        Upload_Date.append("Not mentioned ")  
                
    try: #scraping views
        views=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[4]")
            
        for i in views:
            Views.append(i.text)
    except NoSuchElementException as e:
        Views.append("Not mentioned ")
            
        
    
except NoSuchElementException as s:
    print(s)
    
#creating a dataframe
videos=pd.DataFrame({
"Rank":Rank,
"Name":Name,
"Artist":Artist,
"Upload_Date":Upload_Date,    
"Views":Views
})

#printing dataframe
videos   
     
```

### Q2.  Scrape the details team India’s international fixtures from bcci.tv. Url = https://www.bcci.tv/.
You need to find following details:
#### A) Match title (I.e. 1st ODI) B) Series C) Place D) Date E) Time

```python

from selenium import webdriver
import time
import requests
import shutil
import os
from webdriver_manager.chrome import ChromeDriverManager

option = webdriver.ChromeOptions()
option.headless=True
driver = webdriver.Chrome(ChromeDriverManager().install(), options=option)

url="https://www.bcci.tv/"
driver.get(url)

time.sleep(3)
Title=[]
Series=[]
Place=[]
Date=[]
Time=[]


try:
    # access link of fixtures
    fixture_option=driver.find_element_by_xpath("//li[@class='drop-down__option  ']/a")
    driver.get(fixture_option.get_attribute('href'))
    
except ElementNotInteractableException as n:
        print(n)
        
try:
    #scraping titles
    titles=driver.find_elements_by_xpath("//div[@class='fixture__format-strip']/span[2]")
    for i in titles:
        title=i.text
        Title.append(title)
except NoSuchElementException as e:
    print(e)

try:
    #scraping series
    series=driver.find_elements_by_xpath("//div[@class='fixture__description u-unskewed-text']/p/strong")
    for i in series:
        series_list=i.text
        Series.append(series_list)
except NoSuchElementException as e:
    print(e)  
    
try:
    #scraping place of matches
    place_list=driver.find_elements_by_xpath("//div[@class='fixture__description u-unskewed-text']/p/span")
    for i in place_list:
        place=i.text
        Place.append(place)
except StaleElementReferenceException as e:
    print(e)      
    
try:
    #scraping date of matches
    date_list=driver.find_elements_by_xpath("//div[@class='fixture__description u-unskewed-text']/span/strong")
    for i in date_list:
        date=i.text
        Date.append(date)
except NoSuchElementException as e:
    print(e) 
    
try:
    #scraping time of matches
    time_list=driver.find_elements_by_xpath("//div[@class='fixture__description u-unskewed-text']/span")
    for i in time_list:
        time=i.text.split(' ')
        Time.append(time[3:])
except NoSuchElementException as e:
    print(e) 
    
print(len(Series),len(Title),len(Place),len(Date),len(Time))
#creating a dataframe
fixtures=pd.DataFrame({
"Series":Series,
"Title":Title,
"Place":Place,
"Date":Date,    
"Time":Time
})

#printing dataframe
fixtures  
```

### Q 5 Scrape the details of trending repositories on Github.com. Url = https://github.com/
You have to find the following details:
#### A) Repository title B) Repository description C) Contributors count D) Language used

```python

from selenium import webdriver
import time
import requests
import shutil
import os
from webdriver_manager.chrome import ChromeDriverManager

option = webdriver.ChromeOptions()
option.headless=True
driver = webdriver.Chrome(ChromeDriverManager().install(), options=option)

url="https://github.com/"
driver.get(url)

time.sleep(3)
Repository_Title=[]
Repository_Desc=[]
Language=[]
Contributors=[]


try:
    
    try:
        #accessing link to trending repository
        selenium=driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a")
        driver.get(selenium.get_attribute('href'))
        time.sleep(2)
    
    except ElementNotInteractableException as n:
        print(n)
    
    try:
        #scraping repository title
        titles=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
        for i in titles:
            title=i.text
            Repository_Title.append(title)
    except NoSuchElementException as e:
        Repository_Title.append("Not given")

    try:
        #scraping repository Description
        descs=driver.find_elements_by_xpath("//p[@class='col-9 color-text-secondary my-1 pr-4']")
        for i in descs:
            desc=i.text
            Repository_Desc.append(desc)
    except NoSuchElementException as e:
        Repository_Desc.append("Not given")
    
    try:
        #scraping language used 
        langs=driver.find_elements_by_xpath("//div[@class='f6 color-text-secondary mt-2']/span[1]")
        print(langs)
        for i in langs:
            if i is not None:
                lang=i.text
                print(lang)
                Language.append(lang)
            else:
                Language.append("--")
    except NoSuchElementException as e:
        Language.append("No info")
    
    try:
        #scraping contributors
        contributors=driver.find_elements_by_xpath("//a[2][@class='Link--muted d-inline-block mr-3']")
        for i in contributors:
            cont=i.text
            Contributors.append(cont)
    except NoSuchElementException as e:
        Contributors.append("No info")     

except WebDriverException as w:
    pass


print(len(Repository_Desc),len(Repository_Title),len(Language),len(Contributors))
#creating a dataframe
trending=pd.DataFrame({
"Repository Title":Repository_Title,
"Repository_Description":Repository_Desc,
"Language Used":Language,
"Contributors":Contributors
})

#printing dataframe
trending  
```

```python

from selenium import webdriver
import time
import requests
import shutil
import os
from webdriver_manager.chrome import ChromeDriverManager

option = webdriver.ChromeOptions()
option.headless=True
driver = webdriver.Chrome(ChromeDriverManager().install(), options=option)

url="https://archive.ics.uci.edu/ml/index.php"
driver.get(url)

time.sleep(3)
Name=[]
DataType=[]
Task=[]
AttributeType=[]
NoOfInstance=[]
NoOfAttribute=[]
Year=[]

try:
    
    try:
        #accessing view all datasets
        selenium=driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]/a")
        driver.get(selenium.get_attribute('href'))
        time.sleep(2)
        
        
    except ElementNotInteractableException as n:
        print(n)

    
    try:
        #scraping dataset name
        names=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]")
        for i in names:
            name=i.text
            Name.append(name)
    except NoSuchElementException as e:
        Name.append("Not given")

    try:
        #scraping data type
        datatypes=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]")
        for i in datatypes:
            datatype=i.text
            DataType.append(datatype)
    except NoSuchElementException as e:
        DataType.append("Not given")
    
    try:
        #scraping task
        tasks=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]")
        for i in tasks:
            task=i.text
            Task.append(task)
    except NoSuchElementException as e:
        Task.append("No info")
    
    try:
        #scraping attribute type
        attributes=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]")
        for i in attributes:
            attribute=i.text
            AttributeType.append(attribute)
    except NoSuchElementException as e:
        AttributeType.append("No info")  
    
    try:
        #scraping instances
        instances=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]")
        for i in instances:
            instance=i.text
            NoOfInstance.append(instance)
    except NoSuchElementException as e:
        NoOfInstance.append("--")  
    
    try:
        #scraping no of attribute
        totalattributes=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]")
        for i in totalattributes:
            totalattribute=i.text
            NoOfAttribute.append(totalattribute)
    except NoSuchElementException as e:
        NoOfAttribute.append("--")  
     
    try:
        #scraping year
        years=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]")
        for i in years:
            year=i.text
            Year.append(year)
    except NoSuchElementException as e:
        Year.append("--")     

except WebDriverException as w:
    pass

print(len(Name),len(DataType),len(Task),len(AttributeType),len(NoOfInstance),len(NoOfAttribute),len(Year))

#creating a dataframe
datasets=pd.DataFrame({
"Dataset Name":Name,
"Data Type":DataType,
"Task":Task,
"AttributeType":AttributeType,
"No Of Instance":NoOfInstance ,
"No Of Attribute":NoOfAttribute,
"Year":Year
})

#printing dataframe
datasets  
```
