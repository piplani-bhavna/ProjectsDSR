### Q 5. Scrape data for first 10 companies. Scrape the Company name, Number of salaries, Average salary, Min salary, Max Salary.

```python
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get("https://www.glassdoor.co.in/Salaries/index.htm")

try:
    search_job=driver.find_element_by_xpath("//input[@id='keyword']")
    search_job.send_keys("Data Scientist")
    search_loc=driver.find_element_by_xpath("//input[@id='loc']")
    search_loc.send_keys("Noida")
    search_button=driver.find_element_by_xpath("//div[@class='container-form']/button")
    search_button.click()
    
except NoSuchElementException:
    pass

# scrape Company name, Number of salaries, Average salary, Min salary, Max Salary.
company_name=[]
No_of_salaries=[]
Av_salary=[]
min_salary=[]
max_salary=[]

try:
    
    url=("https://www.glassdoor.co.in/Salaries/new-delhi-data-scientist-salary-SRCH_IL.0,9_IM1083_KO10,24.htm?clickSource=searchBtn")
    driver.get(url)
    
    try:        
        name_tags=driver.find_elements_by_xpath("//div[@data-test='job-info']/p[2]")
        for i in name_tags:
            name=i.text
            company_name.append(name)
    except NoSuchElementException as e:
        pass
    
    number_tags=driver.find_elements_by_xpath("//div[@data-test='job-info']/p[5]")
    for i in number_tags:
        number=i.text
        No_of_salaries.append(number)
        
    avSal_tags=driver.find_elements_by_xpath("//div[@class='col-2 d-none d-md-flex flex-row justify-content-end']/strong")
    for i in avSal_tags:
        avSal=i.text
        Av_salary.append(avSal)
       
    min_tags=driver.find_elements_by_xpath("//div[@class='common__RangeBarStyle__values d-flex justify-content-between ']")
    for i in min_tags:
        salary=i.text.split('\n')
        max_salary.append(salary[1])
        min_salary.append(salary[0])
        
except NoSuchElementException as e:
    pass

#print(len(company_name),len(No_of_salaries),len(Av_salary),len(min_salary),len(max_salary))
#import to dataframe
jobs=pd.DataFrame({})
jobs["Company Name"]=company_name[:10]
jobs["No. of salaries"]=No_of_salaries[:10]
jobs["Average Salary"]=Av_salary[:10]
jobs["Min. Salary"]=min_salary[:10]
jobs["Max. Salary"]=max_salary[:10]
jobs

```

### Q 6. Scrape brand , Product Desc, Price , Discount of first 100 sunglasses listings on flipkart.com.

```python
driver.get("https://www.flipkart.com/")

try:
    search_sunglasses=driver.find_element_by_xpath("//input[@class='_3704LK']")
    search_sunglasses.send_keys("sunglasses")   
    search_button=driver.find_element_by_xpath("//div[@class='_3OO5Xc']/button")
    search_button.click()
    
except NoSuchElementException:
    pass
# scrape brand , Product Desc, Price , Discount
brand_name=[]
Product_Desc=[]
Price=[]
Disc=[]

try:
    
    url=("https://www.flipkart.com/search?q=sunglasses&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
    driver.get(url)
except NoSuchElementException as e:
    pass

for i in range(0,3):
    name_tags=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
    for i in name_tags:
        l=i.text
        brand_name.append(l)
    
    desc_tags=driver.find_elements_by_xpath("//a[@class='IRpwTa']")
    for i in desc_tags:
        desc=i.text
        Product_Desc.append(desc)
        
    price_tags=driver.find_elements_by_xpath("//div[@class='_30jeq3']")
    for i in price_tags:
        price=i.text
        Price.append(price)
        
    disc_tags=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']/span[1]")
    for i in disc_tags:
        disc=i.text
        Disc.append(disc)
        
    # clicking next button to scrape data from 2nd and 3rd page    
    nxt_button=driver.find_element_by_xpath("//a[@class='_1LKTO3']/span[1]")
    if nxt_button.text=='Next':
        nxt_button.click()
        time.sleep(5)


sunglasses_df=pd.DataFrame({'Brand':brand_name[:100],
                         'Description':Product_Desc[:100],
                         'price':Price[:100],
                         'Discount':Disc[:100],
                         })
    
print(sunglasses_df)
```

## Q9. scrape 100 shoes which include “Brand” of the shoes , Short Shoe description, price 

```python
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python

# scrape brand , short Desc, Price 
brand_name=[]
Product_Desc=[]
Price=[]


url=("https://www.myntra.com/shoes")
driver.get(url)

try:
    time.sleep(2)
    label_element = driver.find_element_by_xpath("//*[@id='mountRoot']/div/div[1]/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label/div")
    label_element.click()
    time.sleep(2)
    #TextValue  = label_element.text('Black')
    filter_button=driver.find_element_by_xpath("//*[@id='mountRoot']/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label/div")
    filter_button.click()
    driver.refresh()  
except NoSuchElementException as e:
    pass

for i in range(0,2):
    name_tags=driver.find_elements_by_xpath("//h3[@class='product-brand']")
    #print(name_tags)
    for i in name_tags:
        l=i.text
        #print(l)
        brand_name.append(l)
    
    desc_tags=driver.find_elements_by_xpath("//h4[@class='product-product']")
    #print(desc_tags)
    for i in desc_tags:
        desc=i.text
        #print(desc)
        Product_Desc.append(desc)
    
        
    try:    
        price_tags=driver.find_elements_by_xpath("//div[@class='product-price']/span[1]")
        for i in price_tags[:100]:
            price=i.text
            Price.append(price[:9])
    except NoSuchElementException as e:
        pass
    # clicking next button to scrape data for 100 records    
    nxt_button=driver.find_element_by_xpath("//li[@class='pagination-next']/a")
    if nxt_button.text=='Next':
        nxt_button.click()
        time.sleep(5)
    
    

print(len(brand_name),len(Product_Desc),len(Price))
shoes_df=pd.DataFrame({'Brand':brand_name[:100],
                         'Description':Product_Desc[:100],
                         'price':Price[:100]
                         })
    
print(shoes_df)
```

## Q 10. scrape 3 attributes for each laptop: title , Ratings , Price from amazon.in


```python
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python

# scrape title , ratings, Price 
Title=[]
Ratings=[]
Price=[]


url=("https://www.amazon.in")
driver.get(url)

#locating the search bar
try:
    search_bar=driver.find_element_by_id("twotabsearchtextbox")
    search_bar.clear()
    search_bar.send_keys("laptops")   
    search_button=driver.find_element_by_id("nav-search-submit-button")
    search_button.click()
    
except NoSuchElementException:
    pass

#locating core i7 filter
filter_button=driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']")
for i in filter_button:
    if i.text=='Intel Core i7':
        i.click()
        break
        
#locating core i7 filter
filter_button=driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']")
for i in filter_button:
    if i.text=='Intel Core i9':
        i.click()
        break

#scraping titles
title=driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-normal']")
for i in title[:10]:
    Title.append(i.text)
    
#scraping price
price=driver.find_elements_by_xpath("//span[@class='a-price-whole']")
for i in price[:10]:
    Price.append(i.text)
    
#locating ratings
urls=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
url=[]
for i in urls[:10]:
    url.append(i.get_attribute('href'))
for j in url:
    driver.get(j)
    try:
        rating=driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        rating.click()
        ratings=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        Ratings.append(ratings.text)
    except NoSuchElementException as e:
        Ratings.append("No ratings")
        
print(len(Title),len(Price),len(Ratings))
laptop_df=pd.DataFrame({'Title':Title[:10],
                         'Price':Price[:10],
                         'Ratings':Ratings[:10]
                         })
    
print(laptop_df)
```
